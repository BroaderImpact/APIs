```python
import requests
import os
import pandas as pd
import json
```

```python
api_key = os.environ.get('API_KEY')
```

```python
params = (
    ('q', 'USPS'),
    ('from', '2020-08-15'),
    ('sortBy', 'publishedAt'),
    ('apiKey', api_key),
    ('pageSize', '10'),
    ('page', '1'),
)
```

```python
response = requests.get('https://newsapi.org/v2/everything', params=params)
```

```python
print(response.text)
```

```python
response_json = response.json()
```

```python
print(response_json)
```

```python
df = pd.DataFrame(response_json)
```

```python
df
```

```python
articles = df['articles'].apply(pd.Series)
```

```python
articles
```

```python
new = articles['source'].apply(pd.Series)
```

```python
new_column = new['name']
```

```python
articles.insert(0, "name", new_column, True)
```

```python
articles
```

```python
articles.drop(['source'], axis=1)
```

```python
articles.to_csv('articles.csv')
```
